```python
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
Caracteristicas, Labels = make_blobs(
    n_samples=600,
    centers = 4,
    cluster_std = 2.6,
    random_state=56
)
```

```python
scalar = StandardScaler()
```

```python
scalar_caractristicas=scalar.fit_transform(Caracteristicas)
```

```python
kmeans_valores={
    'init':'random',
    'n_init':10,
    'max_iter':300,
    'random_state':56
}
```

```python
SER=[]
for k in range(1,11):
  kmeansCT=KMeans(n_clusters=k, **kmeans_valores)
  kmeansCT.fit(scalar_caractristicas)
  SER.append(kmeansCT.inertia_)
```

```python
plt.plot(range(1,11), SER)
```

```python
k1=KneeLocator(range(1,11), SER, curve='convex', direction='decreasing')
```

```python
k1.elbow
```

```python
coeficiente_silueta=[]
for k in range(2,11):
  kmeans_S=KMeans(n_clusters=k, **kmeans_valores)
  kmeans_S.fit(scalar_caractristicas)
  score=silhouette_score(scalar_caractristicas, kmeans_S.labels_)
  coeficiente_silueta.append(score)
```

```python
plt.plot(range(2,11), coeficiente_silueta)
```
