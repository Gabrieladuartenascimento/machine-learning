```python
from sklearn.preprocessing import StandardScaler
```

```python
escala = StandardScaler()
```

```python
escala.fit(data.drop('TARGET CLASS', axis=1))
```

```python
escala_dados = escala.transform(data.drop('TARGET CLASS', axis=1))
```

```python
escala_dados
```

```python
data_frame = pd.DataFrame(escala_dados, columns=data.columns[:-1])
```

```python
data_frame
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_Treino, X_Teste, Y_Treino, Y_Teste=train_test_split(escala_dados, data['TARGET CLASS'], test_size=0.3 )
```

```python
X_Treino, Y_Treino
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
Knn = KNeighborsClassifier(n_neighbors=1)
```

```python
Knn.fit(X_Treino, Y_Treino)
```

```python
predicao = Knn.predict(X_Teste)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(Y_Teste, predicao))
```

```python
print(classification_report(Y_Teste, predicao))
```

```python
Taxa_erro = []
for i in range(1,100):
  Knn=KNeighborsClassifier(n_neighbors=i)
  Knn.fit(X_Treino, Y_Treino)
  preditor_i=Knn.predict(X_Teste)
  Taxa_erro.append(np.mean(preditor_i != Y_Teste))
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,100), Taxa_erro)
```

```python
Knn = KNeighborsClassifier(n_neighbors=100)
```

```python
Knn.fit(X_Treino, Y_Treino)
```

```python
preditor = Knn.predict(X_Teste)
```

```python
print(confusion_matrix(Y_Teste, preditor))
```

```python
print(classification_report(Y_Teste, preditor))
```
